## Textblob

Умеет исправлять отдельные слова, есть вероятности слов-коррекций.

```python
!pip install textblob
```

```python
from textblob import TextBlob

misspelled_words = ['wter', 'pthon language', 'moom', 'monly', 'acress whose']
corrected_words = []

for w in misspelled_words:
   text_blob = TextBlob(w)
   text_corrected = text_blob.correct()
   corrected_words.append(text_corrected)

print(', '.join(map(str, corrected_words)))
```

```python
demo_text = "The quikc brown fox jmps ovr the lazy dg."

text_blob = TextBlob(demo_text)
text_corrected = text_blob.correct()
print(text_corrected)
```

```python
from textblob import TextBlob, Word

w = Word('acress')
print(w.spellcheck())
```

## Pyspellchecker

Поддерживает много языков.

```python
!pip install pyspellchecker
```

```python
from spellchecker import SpellChecker

spell = SpellChecker(language='de')

misspelled = spell.unknown(['vile', 'Shvester', 'Hare'])

for word in misspelled:
    print(spell.correction(word))
    print(spell.candidates(word))
```

```python
spell = SpellChecker(language='en')

demo_text = "The quikc brown fox jmps ovr the lazy dg"
misspelled = spell.unknown(demo_text.split())

for word in misspelled:
    print(word)
    print(spell.correction(word))
    print(spell.candidates(word))
    print()
```

```python
spell = SpellChecker(language='ru')

misspelled = spell.unknown(['пчолы', 'любофь'])

for word in misspelled:
    print(spell.correction(word))
    print(spell.candidates(word))
```

## Spylls

Улучшение классического подхода при помощи N-грамм

```python
!pip install spylls
```

```python
from spylls.hunspell import Dictionary

# en_US dictionary is distributed with spylls
# See docs to load other dictionaries
dictionary = Dictionary.from_files('en_US')

print(dictionary.lookup('huney'))
for suggestion in dictionary.suggest('huney'):
    print(suggestion)
```

```python
demo_text = "The quikc brown fox jmps ovr the lazy dg"

for word in demo_text.split():
    for suggestion in dictionary.suggest(word):
        if not dictionary.lookup(word):
            print(word)
            print(suggestion)
            print()
```

## SpaCy

Использует BERT и учитывает контекст.

```python
!pip install spacy
!pip install contextualSpellCheck
```

```python
import spacy
import contextualSpellCheck

nlp = spacy.load('en_core_web_sm')
contextualSpellCheck.add_to_pipe(nlp)
```

```python
for word in ['wter was running in the sink', 'pthon programming language',
             'run away form this man ', 'monly stipend is gived to students', 'acress whose who left messages to you']:
    doc = nlp(word)
    print(doc._.outcome_spellCheck)
```

```python
demo_text = "The quikc brown fox jmps ovr the lazy dg"

doc = nlp(demo_text)
print(doc._.outcome_spellCheck)
```

## Трансформеры

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Bhuvana/t5-base-spellchecker")

model = AutoModelForSeq2SeqLM.from_pretrained("Bhuvana/t5-base-spellchecker")


def correct(inputs):
    input_ids = tokenizer.encode(inputs,return_tensors='pt')
    sample_output = model.generate(
        input_ids,
        do_sample=True,
        max_length=50,
        top_p=0.95,
        num_return_sequences=1
    )
    res = tokenizer.decode(sample_output[0], skip_special_tokens=True)
    return res


text = "christmas is celbrated on decembr 25 evry ear"

print(correct(text))
```

```python
demo_text = "The quikc brown fox jmps ovr the lazy dg"

print(correct(demo_text))
```
